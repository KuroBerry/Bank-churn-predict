```python
import numpy as np
import pandas as pd
```

# Re-train with full dataset

```python
train_data = pd.DataFrame(pd.read_csv('data/train.csv', index_col=['id']))
```

```python
train_data
```

```python
train_data = train_data.drop( 'CustomerId', axis=1)
train_data = train_data.drop( 'Surname', axis=1)
```

```python
# train_data['TenureAgeRatio'] = train_data['Tenure']/train_data['Age']
# train_data = train_data.drop( 'Age', axis=1)
# train_data = train_data.drop('Tenure', axis=1)
```

```python
train_data['ActiveCrCard'] = train_data['HasCrCard']+(train_data['IsActiveMember'])
train_data = train_data.drop( 'HasCrCard', axis=1)
train_data = train_data.drop('IsActiveMember', axis=1)
```

```python
train_data = pd.get_dummies(train_data, columns=['Geography', 'Gender', 'NumOfProducts'])
train_data
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder()

name = list(train_data.columns)
type = train_data.dtypes

for i in range(len(type)):
    if type[i] == 'bool':
        le.fit_transform(train_data[name[i]])
        train_data[name[i]] = le.transform(train_data[name[i]])
        
train_data
```

```python
conditions = [
    (train_data['Age'] <= 28),
    (train_data['Age'] > 28) & (train_data['Age'] <= 33),
    (train_data['Age'] > 33) & (train_data['Age'] <= 40),
    (train_data['Age'] > 40) & (train_data['Age'] <= 50),
    (train_data['Age'] > 50) & (train_data['Age'] <= 69),
    (train_data['Age'] > 69)
]
choices = [1, 2, 3, 4, 5, 6]
train_data['age_slicer'] = np.select(conditions, choices, default=0)

train_data = train_data.drop('Age', axis=1)
print(train_data['age_slicer'])
```

```python
X, y = train_data.drop('Exited', axis=1), train_data['Exited']
```

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

## Train model Catboost

```python
from catboost import CatBoostClassifier 
from joblib import dump, load

modelCatboost = CatBoostClassifier()
modelCatboost.fit(X, y)
```
